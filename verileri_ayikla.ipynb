```python
def drop_columns(df, cols):
    """
    This function removes the columns from a DataFrame.
    """ 
    for col in cols:
        try:
             df.drop([col], axis=1, inplace=True)
        except Exception as e:
            print(f"Bir hata oluştu: {e}")
```

```python
import pandas as pd
import os

# 1. Dosyaları oku ve birleştir
files = os.listdir('hepsiemlak')
df = pd.concat([pd.read_csv(f"hepsiemlak/{file}") for file in files], ignore_index=True)

# 2. Sütun isimlerini düzelt
df.columns = df.columns.str.strip()

# 3. Sütunları yazdır
print("Sütunlar:", df.columns.tolist())

# 4. Adres bilgisini içeren sütunu bul (örnek: 'Adres', 'img-link href', vb.)
adres_column = None
for col in df.columns:
    if df[col].astype(str).str.contains('/').any():
        adres_column = col
        break

if adres_column:
    print(f"Adres bilgisi şu sütunda bulundu: {adres_column}")
    
    # konum bilgilerini ayır
    df['location'] = df[adres_column].astype(str).apply(lambda x: x.split('/')[3] if len(x.split('/')) > 3 else '')
    df['city'] = df['location'].str.split('-').str[0]
    df['district'] = df['location'].str.split('-').str[1]
    df['neighborhood'] = df['location'].str.split('-').str[2]
else:
    print("Adres bilgisi içeren bir sütun bulunamadı. Lütfen manuel kontrol et.")

# 5. Gereksiz sütunları kaldır
drop_cols = [
    'photo-count', 'currency', 'list-view-date', 'left',
    'eids-badge__description', 'he-lazy-image src', 'eids-badge__label',
    'img-wrp href', 'img-wrp href 2', 'he-lazy-image src 3',
    'listing-card--owner-info__name'
]
df.drop([col for col in drop_cols if col in df.columns], axis=1, inplace=True)

# 6. city, district, neighborhood varsa örnekleri göster
if all(col in df.columns for col in ['city', 'district', 'neighborhood']):
    print(df[['city', 'district', 'neighborhood']].sample(50))
else:
    print("Konum sütunları eksik.")
```

```python
def extract_floor(x):
    if isinstance(x, str):
        if 've üzeri' in x:
            return int(x.split(' ')[0])  
        try:
            return int(x.split('.')[0]) 
        except ValueError:
            return 0  
    return 0  

replace_dict = {
    'En Üst Kat': '5.Kat',
    'Giriş Katı': '1.Kat',
    'Bahçe Katı': '0.Kat',
    'Ara kat': '3.Kat',
    'Yüksek Giriş': '1.Kat',
    'Kot 2': '-2.Kat',
    'KOt 1': '-1.Kat',
    'Zemin': '0.Kat',
    'Çatı Katı': '5.Kat',
    'Teras Katı': '5.Kat',
}

df['celly 4'] = df['celly 4'].replace(replace_dict)
df['celly 4'] = df['celly 4'].fillna('0.Kat')
df['floor'] = df['celly 4'].apply(extract_floor)
print(df['floor'].unique())
```

```python
drop_columns(df, ['celly 4'])
print(df.info())
```
